```python
import selenium
import pandas as pd
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException,ElementNotVisibleException
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
from selenium.webdriver.edge.service import Service
```

```python
s = Service("msedgedriver.exe")
```

```python
url =  'https://droom.in/cars/used'
driver = webdriver.Edge(service = s)
driver.get(url)
```

```python
i=0
while(i<400):
      nxtbtn = driver.find_element_by_xpath('/html/body/div[1]/div/div[2]/div/div[3]/div/div[3]/button/span[1]')
      nxtbtn.click()
      time.sleep(5)
      i+=1
```

```python
produrl =[]
```

```python
produrl_tag = driver.find_elements_by_xpath("//a[@class='MuiTypography-root MuiLink-root MuiLink-underlineHover MuiTypography-colorPrimary']")
```

```python
for u in produrl_tag:
 try: 
     produrl.append(u.get_attribute("href"))
 except:
    produrl.append('-')
```

```python
produrl
```

```python
produrl2 = list(set(produrl))
```

```python
len(produrl2)
```

```python
Brand = [] 
model= [] 
cartype= [] 
manufyear= [] 
drivenkilometers= [] 
fuel= [] 
numberowners= [] 
location= []
price= []
transmission= []
seatingcapacity= []
mileage = []
```

```python
for i in produrl2:
    driver.get(i)
    time.sleep(1)
    try:
        brandtag = driver.find_element_by_xpath("/html/body/div[8]/div[1]/div[1]/ol/li[4]/a/span")
        Brand.append(brandtag.text)
    except:
        Brand.append('-')
        
    try:
        fueltag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[6]/div")
        fuel.append(fueltag.text)
    except:
        fuel.append('-')   
    try:
        modeltag = driver.find_element_by_xpath("/html/body/div[8]/div[1]/div[1]/ol/li[5]/span")
        model.append(modeltag.text)
    except:
        model.append('-')
        
    try:
        drivenkilometerstag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[1]/div")
        drivenkilometers.append(drivenkilometerstag.text.split(" ")[0])
    except:
        drivenkilometers.append('-')
        
    try:
        cartypetag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[7]/div")
        cartype.append(cartypetag.text)
    except:
        cartype.append('-')
    try:
        numberownerstag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[3]/div")
        numberowners.append(numberownerstag.text)
    except NoSuchElementException:
        numberowners.append('-')
    try:
        manufyeartag = driver.find_element_by_xpath("/html/body/div[8]/div[1]/div[1]/ol/li[5]/span")
        manufyear.append(manufyeartag.text.split(" ")[-1])
    except:
        manufyear.append('-')
    try:
        pricetag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[2]/div[1]/div[1]/div[5]/div[1]/div[2]")
        price.append(pricetag.text.split(" ")[1])
    except:
        price.append('-')
    
    try:
        locationtag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[2]/div")
        location.append(locationtag.text)
    except:
        location.append('-')
    
    try:
        transmissiontag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[9]/div")
        transmission.append(transmissiontag.text)
    except:
        transmission.append('-')
    try:
        seatingcapacitytag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[8]/div")
        seatingcapacity.append(seatingcapacitytag.text.split(" ")[0])
    except:
        seatingcapacity.append('-')
    try:
        mileagetag = driver.find_element_by_xpath("/html/body/div[8]/div[2]/div[1]/div[1]/div[2]/div/div/ul/li[10]/div")
        mileage.append(mileagetag.text.split(" ")[0])
    except:
        mileage.append('-')
```

```python
Price = []
for p in price:
    Price.append(p.replace(",",""))
    
```

```python
Mileage = []
for m in mileage:
    Mileage.append(m.replace("kmpl",""))
```

```python
UsedCars = {'Brand Name':Brand,'Model Name':model,'Year':manufyear,'Type':cartype,'Total Kilometers Driven':drivenkilometers,'Fuel Type':fuel,'Owner':numberowners,'Location':location,'Transmission':transmission,'Seating Capacity':seatingcapacity,'Mileage':Mileage,'Price':Price}

UsedCarsdf  = pd.DataFrame(UsedCars)
```

```python
UsedCarsdf.to_csv('UsedCarsscrape.csv') # Saving as .csv file
```
