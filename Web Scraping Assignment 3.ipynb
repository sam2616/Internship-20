```python
i=2
while(i>0): 
    nextbtn = driver.find_element_by_xpath("//li[@class='a-last']")
    nextbtn.click()
    time.sleep(3)
    url_tag = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for u in url_tag:
          url.append(u.get_attribute("href"))
    i=i-1
```

```python
Brand = []
prod = []
rate = []
ratenum = []
price = []
ex = []
delv = []
avl = []
det = []
```

```python
for i in url:
    driver.get(i)
    time.sleep(5)
    try:
        brandtag = driver.find_element_by_xpath("//td[@class='a-size-base prodDetAttrValue']")
        Brand.append(brandtag.text)
    except:
        Brand.append('-')
    try:
        prodtag = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        prod.append(prodtag.text)
    except:
        prod.append('-')
    try:
        ratetag = driver.find_element_by_xpath("//div[@id='averageCustomerReviews']//span[@class='a-icon-alt']")
        rate.append(ratetag.get_attribute("innerHTML"))
    except NoSuchElementException:
        rate.append('-')
    try:
        ratenumtag = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        ratenum.append(ratenumtag.get_attribute("innerHTML"))
    except:
        ratenum.append('-')
    try:
        pricetag = driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        price.append(pricetag.text)
        
    except NoSuchElementException:
        try:
            pricetag = driver.find_element_by_xpath("//span[@id='priceblock_dealprice']")
            price.append(pricetag.text)
        except:
            price.append('-')
    
    try:
        extag = driver.find_element_by_xpath("//div[@class='a-accordion-row-a11y']//h5")
        ex.append(extag.text.replace('\n',' ').replace('   ',''))
    except:
        ex.append('-')
    
    try:
        delvtag = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        delv.append(delvtag.text)
    except:
        delv.append('-')
    try:
        avltag = driver.find_element_by_xpath("//div[@id='availability']//span")
        avl.append(avltag.text)
    except:
        avl.append('-')
    try:
        dettag = driver.find_element_by_xpath("//div[@id='feature-bullets']")
        det.append(dettag.text.replace('\nShow More','.').replace('\n','').replace('About this item',""))
    except:
        det.append('-')
```

```python
Laptops  = pd.DataFrame({})
Laptops['Brand Name']=Brand
Laptops['Product Name']=prod
Laptops['Price']=price
Laptops['Rating']=rate
Laptops['Number of Ratings']=ratenum
Laptops['Exchange']=ex
Laptops['Expected Delivery']=delv
Laptops['Availability']=avl
Laptops['Other Details']=det
Laptops['Product Url']=url
```

```python
name = []
dist = []

rev = []
ovrrev = []
pvt = []
dor = []
fac = []
url = []
```

```python
url_tag = driver.find_elements_by_xpath("//a[@class='view-button']")
for u in url_tag:
    url.append(u.get_attribute("href"))
```

```python
name_tag= driver.find_elements_by_xpath("//h2[@class='title title-6']//a")
for n in name_tag:
    name.append(n.text)
dist_tag= driver.find_elements_by_xpath("//a[@class='show-on-map']/span[1]")
for d in dist_tag:
    dist.append(d.text)
rate_tag= driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']/div[1]")
for r in rate_tag:
    rate.append(r.text.replace('\n0 Total Reviews',''))
rev_tag= driver.find_elements_by_xpath("//div[@class='reviews']")
for re in rev_tag:
    rev.append(re.text)
ovrrev_tag= driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']//span")
for o in ovrrev_tag:
    ovrrev.append(o.text)
pvt_tag= driver.find_elements_by_xpath("//a[@class='prices']/div[1]")
for p in pvt_tag:
    pvt.append(p.text.replace("\n",''))
dor_tag= driver.find_elements_by_xpath("//a[@class='prices']/div[2]")
for d in dor_tag:
    dor.append(d.text.replace("\n",''))
fac_tag= driver.find_elements_by_xpath("//div[@class='facilities-label facilities']")
for f in fac_tag:
    fac.append(f.text.replace("\n",' + '))
```

```python
import time 


i=0

while(i<2):
    nextbtn = driver.find_element_by_xpath("//i[@class='core-icon icon-core-chevron-right']")
    nextbtn.click()
    time.sleep(5)
    url_tag = driver.find_elements_by_xpath("//a[@class='view-button']")
    for u in url_tag:
        url.append(u.get_attribute("href"))
    
    name_tag= driver.find_elements_by_xpath("//h2[@class='title title-6']//a")
    for n in name_tag:
        name.append(n.text)
    dist_tag= driver.find_elements_by_xpath("//a[@class='show-on-map']/span[1]")
    for d in dist_tag:
        dist.append(d.text)
    
    rev_tag= driver.find_elements_by_xpath("//div[@class='reviews']")
    for re in rev_tag:
        rev.append(re.text)
    ovrrev_tag= driver.find_elements_by_xpath("//div[@class='rating rating-summary-container big']//span")
    for o in ovrrev_tag:
        ovrrev.append(o.text)
    pvt_tag= driver.find_elements_by_xpath("//a[@class='prices']/div[1]")
    for p in pvt_tag:
        pvt.append(p.text.replace("\n",'').replace("Privates From",""))
    dor_tag= driver.find_elements_by_xpath("//a[@class='prices']/div[2]")
    for d in dor_tag:
        dor.append(d.text.replace("\n",'').replace("Dorms From",""))
    fac_tag= driver.find_elements_by_xpath("//div[@class='facilities-label facilities']")
    for f in fac_tag:
        fac.append(f.text.replace("\n",' + '))
    
    
    i=i+1
```

```python
rate = []
desc = []
for u in url:
    driver.get(u)
    time.sleep(3)
    rate_tag= driver.find_element_by_xpath("//div[@class='keyword']/span")
    rate.append(rate_tag.text)
    desc_tag = driver.find_element_by_xpath("//div[@class='collapsible']//div[2]")
    try:
        desc.append(desc_tag.text.replace("\n"," "))
    except:
        desc.append('-')
    
```
