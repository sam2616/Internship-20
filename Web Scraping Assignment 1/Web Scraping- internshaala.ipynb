```python
import pandas as pd
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://internshala.com/fresher-jobs/")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
print(soup.prettify())
```

```python
job_titles=[]
```

```python
titles=soup.find_all('div',class_="heading_4_5 profile")
```

```python
for i in titles:
    job_titles.append(i.text.replace("\n",""))
```

```python
job_titles
```

```python
company_name=[]
```

```python
company=soup.find_all('a',class_="link_display_like_text")
```

```python
for i in company:
    company_name.append(i.text.strip())
```

```python
company_name
```

```python
other_info=[]
```

```python
other_details=soup.find_all('div',class_="item_body",id=False)
```

```python
for i in other_details:
    other_info.append(i.text.strip())
```

```python
other_info
```

```python
CTC=other_info[0::2]
```

```python
apply_date=other_info[1::2]
```

```python
jobs=pd.DataFrame({})
```

```python
print (len(job_titles),len(company_name),len(CTC),len(apply_date))
```

```python
jobs['Job Title']=job_titles
jobs['Company Name']=company_name
jobs['CTC']=CTC
jobs['Apply Date']=apply_date
```
