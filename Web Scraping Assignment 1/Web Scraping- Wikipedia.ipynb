```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
titles=soup.find_all(['h1','h2','h3','h4','h5','h6'])
```

```python
titles
```

```python
print("Header Tags",*titles,sep="\n\n")
```
