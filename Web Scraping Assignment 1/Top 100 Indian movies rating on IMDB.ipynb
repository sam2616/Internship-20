```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
page=requests.get("https://www.imdb.com/india/top-rated-indian-movies/?pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=8a7876cd-2844-4017-846a-2c0876945b7b&pf_rd_r=4G8GNT6KYNCQS5M09AT7&pf_rd_s=right-5&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_india_tr_rhs_1")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
rating=soup.find_all('td',class_="ratingColumn imdbRating")
```

```python
IMDB_rating=[]
```

```python
for i in rating:
    IMDB_rating.append(i.text.replace("\n",""))
```

```python
IMDB_rating
```

```python
yor=soup.find_all('span',class_="secondaryInfo")
```

```python
YOR=[]
```

```python
for i in yor:
    YOR.append(i.text)
```

```python
YOR
```

```python
titles=soup.find_all('td',class_='titleColumn')
```

```python
Titles=[]
```

```python
titles
```

```python
for i in titles:
    for j in i.find_all('a',href=True):
        Titles.append(j.text)
```

```python
Titles
```

```python
movies=pd.DataFrame({})
```

```python
movies['Name']=Titles
movies['IMDB rating']=IMDB_rating
movies['Year of release']=YOR
```

```python
movies
```

```python
movies.index=movies.index+1
```

```python
movies
```

```python
Top_100_Indian_movies=movies[0:100]
```
