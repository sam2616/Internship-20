```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
Other_info=[]
```

```python
other_info=soup.find_all('div',class_="font-semi-bold heading-6")
```

```python
for i in other_info:
    Other_info.append(i.text)
```

```python
Other_info
```

```python
Area=Other_info[0::3]
```

```python
Area
```

```python
Emi=Other_info[1::3]
```

```python
Price=Other_info[2::3]
```

```python
Location=[]
```

```python
location=soup.find_all('div',class_="nb__2CMjv")
```

```python
for i in location:
    Location.append(i.text.replace('Independent House',"").replace(',', " ").strip())
```

```python
Location
```

```python
Title=[]
```

```python
title=soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY")
```

```python
title
```

```python
for i in title:
    Title.append(i.text)
```

```python
Title
```

```python
House_details=pd.DataFrame({})
```

```python
House_details['House Title']=Title
House_details['Location']=Location
House_details['Area']=Area
House_details['EMI']=Emi
House_details['Price']=Price
```
