```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
page=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
rating=soup.find_all('td',class_="ratingColumn imdbRating")
```

```python
IMDB_rating=[]
```

```python
for i in rating:
    IMDB_rating.append(i.text.replace("\n",""))
```

```python
year=soup.find_all('span',class_="secondaryInfo")
```

```python
YOR=[]
```

```python
for j in year:
    YOR.append(j.text)
```

```python
titles=soup.find_all('td',class_='titleColumn')
```

```python
movie_name=[]
```

```python
for i in titles:
    for j in i.find_all('a',href=True):
        movie_name.append(j.text)
```

```python
movies=pd.DataFrame({})
```

```python
movies['Name']=movie_name
movies['IMDB rating']=IMDB_rating
movies['Year of Release']=YOR
```

```python
Top_100_movies=movies[0:100]
```

```python
Top_100_movies.index.name='Ranking'
```

```python
Top_100_movies.index=Top_100_movies.index+1
```
