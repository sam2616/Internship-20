```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url="https://bookpage.com"
```

```python
url="https://bookpage.com/reviews"
```

```python
page=requests.get(url)
```

```python
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
print(soup.prettify())
```

```python
bookname=soup.find_all('h4',class_="italic")
```

```python
Bookname=[]
```

```python
for i in bookname:
    Bookname.append(i.text.replace("\n",""))
```

```python
author_name=[]
```

```python
author=soup.find_all('p',class_="sans bold")
```

```python
for i in author:
    author_name.append(i.text.replace("\n",""))
```

```python
Genre=[]
```

```python
genre=soup.find_all('p',class_="genre-links hidden-phone")
```

```python
for i in genre:
    Genre.append(i.text.replace("\n",""))
```

```python
reviews=soup.find_all('div',class_="read-full")
```

```python
Review=[]
```

```python
for i in reviews:
    for j in i.find_all('a',href=True):
        Review.append(base_url+j['href'])
```

```python
import pandas as pd
```

```python
Bookpage=pd.DataFrame({})
```

```python
Bookpage['Book Name']=Bookname
Bookpage['Author Name']=author_name
Bookpage['Genre']=Genre
Bookpage['Book Review']=Review
```

```python
Bookpage=Bookpage[0:5]
```
